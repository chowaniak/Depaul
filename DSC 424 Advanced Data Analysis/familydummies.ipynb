```python
#CSC424 create dummies

## Import Modules
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
##get working directory
os.getcwd()
```

```python
##load famcleaned2.csv via read_csv
fam_df = pd.read_csv('C:/Depaul_Win7/DSC 424 Advanced Data Analysis/Project/family/famcat.csv', sep=",")
fam_df = fam_df.drop(fam_df.columns[0],axis=1)
fam_df.head()
```

```python
fam_cat = fam_df.astype('category')
```

```python
fam_cat.dtypes
```

```python
# create dummy variables k-1
fam_dummy = pd.get_dummies(fam_cat, drop_first=True)
fam_dummy.head()
```

```python
##load famnum via read_csv
fam_num = pd.read_csv('C:/Depaul_Win7/DSC 424 Advanced Data Analysis/Project/family/famnum.csv', sep=",")
fam_num = fam_num.drop(fam_num.columns[0],axis=1)
fam_num.head()
```

```python
#merge data frames
famreg = pd.concat([fam_num, fam_dummy], axis = 1, join_axes=[fam_num.index])
famreg.head()
```

```python
#save dummy variables to csv file
famreg.to_csv('fam_all_dummies.csv', sep=",")
```
